```python
from knowledge_neurons import KnowledgeNeurons, initialize_model_and_tokenizer, model_type
import random

# first initialize some hyperparameters
MODEL_NAME = "gpt2"

# to find the knowledge neurons, we need the same 'facts' expressed in multiple different ways, and a ground truth
TEXTS = [
    "Sarah was visiting [MASK], the capital of france",
    "The capital of france is [MASK]",
    "[MASK] is the capital of france",
    "France's capital [MASK] is a hotspot for romantic vacations",
    "The eiffel tower is situated in [MASK]",
    "[MASK] is the most populous city in france",
    "[MASK], france's capital, is one of the most popular tourist destinations in the world",
]
TEXT = TEXTS[0]
GROUND_TRUTH = "paris"

# these are some hyperparameters for the integrated gradients step
BATCH_SIZE = 20
STEPS = 20 # number of steps in the integrated grad calculation
ADAPTIVE_THRESHOLD = 0.3 # in the paper, they find the threshold value `t` by multiplying the max attribution score by some float - this is that float.
P = 0.5 # the threshold for the sharing percentage

# setup model & tokenizer
model, tokenizer = initialize_model_and_tokenizer(MODEL_NAME)

# initialize the knowledge neuron wrapper with your model, tokenizer and a string expressing the type of your model ('gpt2' / 'gpt_neo' / 'bert')
kn = KnowledgeNeurons(model, tokenizer, model_type=model_type(MODEL_NAME))

# use the integrated gradients technique to find some refined neurons for your set of prompts
refined_neurons = kn.get_refined_neurons(
    TEXTS,
    GROUND_TRUTH,
    p=P,
    batch_size=BATCH_SIZE,
    steps=STEPS,
    coarse_adaptive_threshold=ADAPTIVE_THRESHOLD,
)

# suppress the activations at the refined neurons + test the effect on a relevant prompt
# 'results_dict' is a dictionary containing the probability of the ground truth being generated before + after modification, as well as other info
# 'unpatch_fn' is a function you can use to undo the activation suppression in the model. 
# By default, the suppression is removed at the end of any function that applies a patch, but you can set 'undo_modification=False', 
# run your own experiments with the activations / weights still modified, then run 'unpatch_fn' to undo the modifications
results_dict, unpatch_fn = kn.suppress_knowledge(
    TEXT, GROUND_TRUTH, refined_neurons
)

# suppress the activations at the refined neurons + test the effect on an unrelated prompt
results_dict, unpatch_fn = kn.suppress_knowledge(
    "[MASK] is the official language of the solomon islands",
    "english",
    refined_neurons,
)

# enhance the activations at the refined neurons + test the effect on a relevant prompt
results_dict, unpatch_fn = kn.enhance_knowledge(TEXT, GROUND_TRUTH, refined_neurons)

# erase the weights of the output ff layer at the refined neurons (replacing them with zeros) + test the effect
results_dict, unpatch_fn = kn.erase_knowledge(
    TEXT, refined_neurons, target=GROUND_TRUTH, erase_value="zero"
)

# erase the weights of the output ff layer at the refined neurons (replacing them with an unk token) + test the effect
results_dict, unpatch_fn = kn.erase_knowledge(
    TEXT, refined_neurons, target=GROUND_TRUTH, erase_value="unk"
)

# edit the weights of the output ff layer at the refined neurons (replacing them with the word embedding of 'target') + test the effect
# we can make the model think the capital of france is London!
results_dict, unpatch_fn = kn.edit_knowledge(
    TEXT, target="london", neurons=refined_neurons
)
```

```python
def scaled_input(emb, batch_size, num_batch):
    # emb: (1, ffn_size)
    baseline = torch.zeros_like(emb)  # (1, ffn_size)

    num_points = batch_size * num_batch
    step = (emb - baseline) / num_points  # (1, ffn_size)

    res = torch.cat([torch.add(baseline, step * i) for i in range(num_points)], dim=0)  # (num_points, ffn_size)
    return res, step[0]
```

```python
data_path = '/home/jtj/probing_llama/datasets/security'
safe_file = os.path.join(data_path, 'safe.txt')
unsafe_file = os.path.join(data_path, 'unsafe.txt')
safe_questions = pd.read_csv(safe_file, sep='#', header=None)
unsafe_questions = pd.read_csv(unsafe_file, sep='#', header=None)

probing_layer_idx = 31
safe_mlp = []
safe_attention = []
safe_layer_outputs = []
unsafe_mlp = []
unsafe_attention = []
unsafe_layer_outputs = []

for i, question in enumerate(safe_questions[0]):
    print(i, question)
    attention_outputs = []
    mlp_outputs = []
    layer_outputs_outputs = []
    
    prompt = llama_v2_prompt([{'role': 'user', 'content': question}])
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(inputs.input_ids.cuda(), max_new_tokens = 100, output_hidden_states = True, output_attentions = True, return_dict_in_generate = True)
    text = tokenizer.decode(outputs[0][0], skip_special_tokens = True)
    
    safe_mlp.append(mlp_outputs[probing_layer_idx][0][-1])
    safe_attention.append(attention_outputs[probing_layer_idx][0][0][-1])
    safe_layer_outputs.append(layer_outputs_outputs[probing_layer_idx][0][0][-1])

for i, question in enumerate(unsafe_questions[0]):
    print(i, question)
    attention_outputs = []
    mlp_outputs = []
    layer_outputs_outputs = []
    
    prompt = llama_v2_prompt([{'role': 'user', 'content': question}])
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(inputs.input_ids.cuda(), max_new_tokens = 100, output_hidden_states = True, output_attentions = True, return_dict_in_generate = True)
    text = tokenizer.decode(outputs[0][0], skip_special_tokens = True)
    
    unsafe_mlp.append(mlp_outputs[probing_layer_idx][0][-1])
    unsafe_attention.append(attention_outputs[probing_layer_idx][0][0][-1])
    unsafe_layer_outputs.append(layer_outputs_outputs[probing_layer_idx][0][0][-1])
```

```python
safe_mlp_diff = []
unsafe_mlp_diff = []
safe_unsafe_mlp_diff = []

safe_attention_diff = []
unsafe_attention_diff = []
safe_unsafe_attention_diff = []

safe_layer_outputs_diff = []
unsafe_layer_outputs_diff = []
safe_unsafe_layer_outputs_diff = []

for i in range(len(safe_mlp)):
    for j in range(i+1, len(safe_mlp)):
        safe_mlp_diff.append((safe_mlp[i] - safe_mlp[j]).cpu())
        safe_attention_diff.append((safe_attention[i] - safe_attention[j]).cpu())
        safe_layer_outputs_diff.append((safe_layer_outputs[i] - safe_layer_outputs[j]).cpu())

        unsafe_mlp_diff.append((unsafe_mlp[i] - unsafe_mlp[j]).cpu())
        unsafe_attention_diff.append((unsafe_attention[i] - unsafe_attention[j]).cpu())
        unsafe_layer_outputs_diff.append((unsafe_layer_outputs[i] - unsafe_layer_outputs[j]).cpu())

for i in range(len(safe_mlp)):
#     for j in range(len(unsafe_mlp)):
    safe_unsafe_mlp_diff.append((safe_mlp[i] - unsafe_mlp[i]).cpu())
    safe_unsafe_attention_diff.append((safe_attention[i] - unsafe_attention[i]).cpu())
    safe_unsafe_layer_outputs_diff.append((safe_layer_outputs[i] - unsafe_layer_outputs[i]).cpu())
```

```python
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

all_vectors = np.concatenate((safe_mlp_diff, unsafe_mlp_diff, safe_unsafe_mlp_diff), axis=0)
pca = PCA(n_components=50)
pca_result = pca.fit_transform(all_vectors)
tsne = TSNE(n_components=2, perplexity=30.0)
tsne_result = tsne.fit_transform(pca_result)

labels = ['safe'] * len(safe_mlp_diff) + ['unsafe'] * len(unsafe_mlp_diff) + ['safe_unsafe'] * len(safe_unsafe_mlp_diff)
df = pd.DataFrame(data=tsne_result, columns=['Component 1', 'Component 2'])
df['Category'] = labels

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Component 1', y='Component 2', hue='Category', palette='Set1', s=50)
plt.title('TSNE Scatter Plot of X1, X2, and X3')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Category')
plt.grid(True)
# plt.show()
plt.savefig('results/layer_%s_mlp.png' % probing_layer_idx)

all_vectors = np.concatenate((safe_attention_diff, unsafe_attention_diff, safe_unsafe_attention_diff), axis=0)
pca = PCA(n_components=50)
pca_result = pca.fit_transform(all_vectors)
tsne = TSNE(n_components=2, perplexity=30.0)
tsne_result = tsne.fit_transform(pca_result)

labels = ['safe'] * len(safe_attention_diff) + ['unsafe'] * len(unsafe_attention_diff) + ['safe_unsafe'] * len(safe_unsafe_attention_diff)
df = pd.DataFrame(data=tsne_result, columns=['Component 1', 'Component 2'])
df['Category'] = labels

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Component 1', y='Component 2', hue='Category', palette='Set1', s=50)
plt.title('TSNE Scatter Plot of X1, X2, and X3')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Category')
plt.grid(True)
# plt.show()
plt.savefig('results/layer_%s_attention.png' % probing_layer_idx)

all_vectors = np.concatenate((safe_layer_outputs_diff, unsafe_layer_outputs_diff, safe_unsafe_layer_outputs_diff), axis=0)
pca = PCA(n_components=50)
pca_result = pca.fit_transform(all_vectors)
tsne = TSNE(n_components=2, perplexity=30.0)
tsne_result = tsne.fit_transform(pca_result)

labels = ['safe'] * len(safe_layer_outputs_diff) + ['unsafe'] * len(unsafe_layer_outputs_diff) + ['safe_unsafe'] * len(safe_unsafe_layer_outputs_diff)
df = pd.DataFrame(data=tsne_result, columns=['Component 1', 'Component 2'])
df['Category'] = labels

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Component 1', y='Component 2', hue='Category', palette='Set1', s=50)
plt.title('TSNE Scatter Plot of X1, X2, and X3')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Category')
plt.grid(True)
# plt.show()
plt.savefig('results/layer_%s_layer_outputs.png' % probing_layer_idx)
```
